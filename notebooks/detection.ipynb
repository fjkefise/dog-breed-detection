```python
from ultralytics import YOLO
import pandas as pd
import os
import shutil

# Создаём папку для сохранения моделей
os.makedirs('models/detection', exist_ok=True)

# Загружаем все три модели
models = {
    'yolo11n_20_epochs': YOLO('runs/detect/yolo11n_20_epochs/weights/best.pt'),
    'yolo11n_30_epochs': YOLO('runs/detect/yolo11n_30_epochs/weights/best.pt'),
    'yolo11s_30_epochs': YOLO('runs/detect/yolo11s_30_epochs/weights/best.pt')
}

# Тестируем все модели
results = {}
print("🧪 ТЕСТИРОВАНИЕ ВСЕХ МОДЕЛЕЙ...\n")

for name, model in models.items():
    print(f"Тестируем {name}...")
    metrics = model.val(
        data='dataset_windows.yaml',
        split='val',
        imgsz=256,
        batch=8,
        conf=0.25,
        iou=0.45,
        device='cuda',
        verbose=False
    )
    
    results[name] = {
        'mAP50': metrics.box.map50,
        'mAP50-95': metrics.box.map,
        'Precision': metrics.box.mp,
        'Recall': metrics.box.mr,
        'Speed_ms': sum(metrics.speed.values())
    }
    
    print(f"  mAP50: {metrics.box.map50:.4f}, mAP50-95: {metrics.box.map:.4f}")

# Создаем таблицу сравнения
print("\n" + "="*70)
print("СРАВНИТЕЛЬНАЯ ТАБЛИЦА МОДЕЛЕЙ")
print("="*70)

comparison_data = []
for name, res in results.items():
    comparison_data.append({
        'Модель': name,
        'mAP50': f"{res['mAP50']:.4f}",
        'mAP50-95': f"{res['mAP50-95']:.4f}",
        'Precision': f"{res['Precision']:.4f}",
        'Recall': f"{res['Recall']:.4f}",
        'Скорость (ms)': f"{res['Speed_ms']:.1f}"
    })

df = pd.DataFrame(comparison_data)
print(df.to_string(index=False))

# Экспорт всех моделей
print("\n" + "="*70)
print("ЭКСПОРТ МОДЕЛЕЙ")
print("="*70)

for name, model in models.items():
    print(f"\nЭкспорт {name}...")
    
    # Экспорт в ONNX
    model.export(format='onnx')
    
    # Копируем лучшие веса
    weight_path = f'runs/detect/{name}/weights/best.pt'
    onnx_path = f'runs/detect/{name}/weights/best.onnx'
    
    if os.path.exists(weight_path):
        shutil.copy2(weight_path, f'models/detection/{name}.pt')
        print(f"  ✅ Веса: models/detection/{name}.pt")
    
    if os.path.exists(onnx_path):
        shutil.copy2(onnx_path, f'models/detection/{name}.onnx')
        print(f"  ✅ ONNX: models/detection/{name}.onnx")

# Выбор лучшей модели
print("\n" + "="*70)
print("РЕКОМЕНДАЦИИ ПО ВЫБОРУ МОДЕЛИ")
print("="*70)

best_overall = max(results.items(), key=lambda x: x[1]['mAP50'])
best_detail = max(results.items(), key=lambda x: x[1]['mAP50-95'])
fastest = min(results.items(), key=lambda x: x[1]['Speed_ms'])

print(f"""
🏆 ЛУЧШАЯ ПО ОБЩЕЙ ТОЧНОСТИ (mAP50):
   {best_overall[0]} - {best_overall[1]['mAP50']:.4f}

🎯 ЛУЧШАЯ ПО ДЕТАЛЬНОЙ ТОЧНОСТИ (mAP50-95):
   {best_detail[0]} - {best_detail[1]['mAP50-95']:.4f}

⚡ САМАЯ БЫСТРАЯ:
   {fastest[0]} - {fastest[1]['Speed_ms']:.1f} ms/изображение

📊 РЕКОМЕНДАЦИИ:
1. Для API/Production: {best_overall[0]}
2. Для детальной детекции: {best_detail[0]}
3. Для real-time: {fastest[0]}
4. Для edge-устройств: yolo11n_30_epochs (баланс скорости и точности)
""")

# Создаём итоговый отчет
print("\n" + "="*70)
print("ИТОГОВЫЙ ОТЧЕТ СОХРАНЕН")
print("="*70)

report = f"""
ОТЧЕТ ПО ОБУЧЕНИЮ МОДЕЛЕЙ ДЕТЕКЦИИ СОБАК
Дата: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}

ИСПЫТАННЫЕ МОДЕЛИ:
1. yolo11n_20_epochs - YOLOv11n, 20 эпох
2. yolo11n_30_epochs - YOLOv11n, 30 эпох  
3. yolo11s_30_epochs - YOLOv11s, 30 эпох

РЕЗУЛЬТАТЫ:
{yolo11n_20_epochs}: mAP50={results['yolo11n_20_epochs']['mAP50']:.4f}, mAP50-95={results['yolo11n_20_epochs']['mAP50-95']:.4f}
{yolo11n_30_epochs}: mAP50={results['yolo11n_30_epochs']['mAP50']:.4f}, mAP50-95={results['yolo11n_30_epochs']['mAP50-95']:.4f}
{yolo11s_30_epochs}: mAP50={results['yolo11s_30_epochs']['mAP50']:.4f}, mAP50-95={results['yolo11s_30_epochs']['mAP50-95']:.4f}

ВЫВОДЫ:
1. Все модели показали хорошие результаты (mAP50 > 0.86)
2. YOLOv11n_30_epochs - лучший баланс скорости и точности
3. YOLOv11s_30_epochs - максимальная точность на Japanese_spaniel
4. Разница между 20 и 30 эпохами для YOLOv11n незначительна

МОДЕЛИ СОХРАНЕНЫ В:
• models/detection/yolo11n_20_epochs.pt/.onnx
• models/detection/yolo11n_30_epochs.pt/.onnx  
• models/detection/yolo11s_30_epochs.pt/.onnx

ДЛЯ ИСПОЛЬЗОВАНИЯ В API:
from api.models.detector import DogDetector

# Для скорости
detector = DogDetector(model_path='models/detection/yolo11n_30_epochs.pt')

# Для максимальной точности  
detector = DogDetector(model_path='models/detection/yolo11s_30_epochs.pt')
"""

with open('models/detection/training_report.txt', 'w', encoding='utf-8') as f:
    f.write(report)

print("✅ Отчёт сохранен: models/detection/training_report.txt")
print("✅ Все модели экспортированы в models/detection/")
print("✅ Готово к использованию в API!")
```

```python
def simplify_onnx_model(input_path, output_path):
    model = onnx.load(input_path)
    model_simp, check = onnxsim.simplify(model)
    onnx.save(model_simp, output_path)

def dynamic_quantization(input_model_path, output_model_path):
    quantize_dynamic(
        model_input=input_model_path,
        model_output=output_model_path,
        weight_type=QuantType.QUInt8,
        per_channel=True,
        reduce_range=True
    )
```

```python
import os
from ultralytics import YOLO
import onnx
import onnxsim
from onnxruntime.quantization import quantize_dynamic, QuantType

# 1. Правильные пути
pt_path = "runs/detect/yolo11n_30_epochs/weights/best.pt"
onnx_dir = "models/detection"
os.makedirs(onnx_dir, exist_ok=True)

# 2. Экспорт с imgsz=256 (как обучали) и без simplify
model = YOLO(pt_path)
model.export(format='onnx', imgsz=256, opset=12, simplify=False)

# 3. Находим экспортированный файл (YOLO может сохранить в разные места)
possible_sources = [
    pt_path.replace('.pt', '.onnx'),
    "yolo11n_30_epochs.onnx",
    os.path.join("runs/detect/yolo11n_30_epochs/weights", "best.onnx")
]

source_onnx = None
for path in possible_sources:
    if os.path.exists(path):
        source_onnx = path
        break

if not source_onnx:
    print("Ищем файлы .onnx в проекте:")
    for root, dirs, files in os.walk("."):
        for file in files:
            if file.endswith('.onnx'):
                print(f"  - {os.path.join(root, file)}")
    raise FileNotFoundError("Не найден экспортированный .onnx файл")

# 4. Пути для конечных файлов
base_onnx = os.path.join(onnx_dir, "yolo11n_30_epochs.onnx")
simp_onnx = os.path.join(onnx_dir, "yolo11n_30_epochs_simplified.onnx")
quant_onnx = os.path.join(onnx_dir, "yolo11n_30_epochs_quantized.onnx")

# 5. Перемещаем и переименовываем
import shutil
shutil.move(source_onnx, base_onnx)
print(f"✓ Модель сохранена: {base_onnx}")

# 6. Упрощение
print("Упрощение модели...")
model = onnx.load(base_onnx)
model_simp, check = onnxsim.simplify(model)
onnx.save(model_simp, simp_onnx)
print(f"✓ Упрощенная модель: {simp_onnx}")

# 7. Квантование
print("Квантование модели...")
quantize_dynamic(
    model_input=simp_onnx,
    model_output=quant_onnx,
    weight_type=QuantType.QUInt8,
    per_channel=True,
    reduce_range=True
)
print(f"✓ Квантованная модель: {quant_onnx}")

print("\n✅ Все операции завершены!")
print(f"   Исходная:   {os.path.getsize(base_onnx)/1024/1024:.1f} MB")
print(f"   Упрощенная: {os.path.getsize(simp_onnx)/1024/1024:.1f} MB")
print(f"   Квантованная: {os.path.getsize(quant_onnx)/1024/1024:.1f} MB")
```
